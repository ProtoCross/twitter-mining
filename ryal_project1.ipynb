## Duck Word Cloud

```python
import matplotlib.pyplot as plt
import pickle
from wordcloud import WordCloud

#Load pickle
with open('duck.pkl', 'rb') as f:
    tweets = pickle.load(f)

#Filter out urls
lexemes = []
for tweet in tweets:
    lexemes.extend(tweet.text.split())
for x in lexemes:
    if "https" in x:
        lexemes.remove(x)
#Change to string for compatability with word cloud
text = ' '. join(lexemes)

#Generate word cloud
wordcloud = WordCloud(max_font_size = 40).generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

## Duck Sentiment Analysis

```python
import pickle
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import nltk
nltk.download('vader_lexicon')

from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

with open('duck.pkl', 'rb') as x:
    duck = pickle.load(x)

def plottable_sentiments(tweets):
    texts = [tweet.text for tweet in tweets]
    sents = [analyzer.polarity_scores(t) for t in texts]
    pos_x = [s['pos'] for s in sents]
    neg_y = [s['neg'] for s in sents]
    comp_z = [s['compound'] for s in sents]
    return pos_x, neg_y, comp_z

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
rbx, rby, rbz = plottable_sentiments(duck)

ax.scatter(rbx, rby, rbz, c = 'r', marker = 'o')
ax.set_xlabel('Positive')
ax.set_ylabel('Negative')
ax.set_zlabel('Compound')
plt.show()
```
