## Duck Word Cloud

```python
with open('duck.pkl', 'rb') as f:
    tweets = pickle.load(f)
    
lexemes = []
for tweet in tweets:
    lexemes.extend(tweet.text.split())
for x in lexemes:
    if "https" in x:
        lexemes.remove(x)
text = ' '. join(lexemes)

wordcloud = WordCloud(max_font_size = 40).generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

## Leffen Word Cloud

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

##Load pickle
with open('leffen.pkl', 'rb') as f:
    tweets = pickle.load(f)

#Filter out urls
lexemes = []
for tweet in tweets:
    lexemes.extend(tweet.text.split())
for x in lexemes:
    if "https" in x:
        lexemes.remove(x)
#Change to string for compatability with word cloud
text = ' '. join(lexemes)

#Generate word cloud
wordcloud = WordCloud(max_font_size = 40).generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```
